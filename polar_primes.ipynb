```python
import numpy as np
import pandas as pd
import math

from bokeh.plotting import output_notebook, show, figure
from bokeh.io import curdoc

file = open("primes.csv", "r")
primes = []
for line in file:
    primes.extend(list(map(int, line.split())))
file.close()

output_notebook()
curdoc().theme = 'dark_minimal'

plot = figure(title="First " + str(len(primes)) + " Polar Primes (Scroll to zoom!)", x_range=(-500, 500), y_range=(-500, 500))
plot.circle(list(map(lambda x: x * math.cos(x), primes)), list(map(lambda y: y * math.sin(y), primes)))
plot.title.text_font_size = "2em"
show(plot)
```
